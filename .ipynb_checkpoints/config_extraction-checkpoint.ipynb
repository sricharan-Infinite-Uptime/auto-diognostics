```python
connection = psycopg2.connect(user="iuprod",
                                          password="pratiti",
                                          host="10.0.1.253",
                                          port="5432",
                                          database="iu_admin")

cursor = connection.cursor()
        
query = f"""
    select id, specs, orientation_code from public.monitors where deactivation_date is null and device_identifier is not null order by id;
    """
cursor.execute(query)
records = cursor.fetchall()
result = []

for row in records:
    data = {}
    data['monitor_id'] = row[0]
    data['config'] = row[1]
    data['orientation'] = row[2]   
    result.append(data)

config_prod_df=pd.DataFrame(result)
config_prod_df.shape


#retrieve data from configuration database for speed, bearing type, thresholds and speed range
speed=[]
brg_type=[]
speed_type=[]
speed_range=[]
threshold_X=[]
threshold_Y=[]
threshold_Z=[]
acceleration_=[]
temperature_=[]
audio_=[]
lor=[]
fmax=[]
fmin=[]

for i in range(len(config_prod_df)):
    config_msg=yaml.load(config_prod_df.config[i],Loader=yaml.FullLoader)
    if config_msg is None:
        speed.append(0)
        brg_type.append(0)
        speed_type.append('NA')
        speed_range.append({})
        threshold_X.append([0,0,0])
        threshold_Y.append([0,0,0])
        threshold_Z.append([0,0,0])
        acceleration_.append([0,0,0])
        temperature_.append([0,0,0])
        audio_.append([0,0,0])
        lor.append([0,0,0])
        fmax.append([0,0,0])
        fmin.append([0,0,0])
        
    else:
        speed.append(config_msg['additional_attrs']['rotating_speed'])
        brg_type.append(config_msg['additional_attrs']['bearing_type'])

        if 'rotating_speed_type' in config_msg['additional_attrs']:
            speed_type.append(config_msg['additional_attrs']['rotating_speed_type'])
        else:
            speed_type.append('NA')

        if 'rotating_speed_range' in config_msg['additional_attrs']:
            speed_range.append(config_msg['additional_attrs']['rotating_speed_range'])
        else:
            speed_range.append({})
            
        lor.append(config_msg['spectral_monitoring']['lines_of_resolution'])    
        fmax.append(config_msg['spectral_monitoring']['maximum_frequency'])  
        if 'minimum_frequency' in config_msg['spectral_monitoring']:
            fmin.append(config_msg['spectral_monitoring']['minimum_frequency']) 
        else:
            fmin.append([0,0,0])
        
        
        acceleration_.append([config_msg['conditions']['Total Acceleration']['Warning']['threshold_value'],
                 config_msg['conditions']['Total Acceleration']['Caution']['threshold_value'],
                 config_msg['conditions']['Total Acceleration']['Operational']['threshold_value']])
        
        temperature_.append([config_msg['conditions']['Temperature']['Warning']['threshold_value'],
                 config_msg['conditions']['Temperature']['Caution']['threshold_value'],
                 config_msg['conditions']['Temperature']['Operational']['threshold_value']])
        
        audio_.append([config_msg['conditions']['Audio']['Warning']['threshold_value'],
                 config_msg['conditions']['Audio']['Caution']['threshold_value'],
                 config_msg['conditions']['Audio']['Operational']['threshold_value']])

        threshold_X.append([config_msg['conditions']['X-axis Velocity']['Warning']['threshold_value'],
                 config_msg['conditions']['X-axis Velocity']['Caution']['threshold_value'],
                 config_msg['conditions']['X-axis Velocity']['Operational']['threshold_value']])
        threshold_Y.append([config_msg['conditions']['Y-axis Velocity']['Warning']['threshold_value'],
                 config_msg['conditions']['Y-axis Velocity']['Caution']['threshold_value'],
                 config_msg['conditions']['Y-axis Velocity']['Operational']['threshold_value']])
        threshold_Z.append([config_msg['conditions']['Z-axis Velocity']['Warning']['threshold_value'],
                 config_msg['conditions']['Z-axis Velocity']['Caution']['threshold_value'],
                 config_msg['conditions']['Z-axis Velocity']['Operational']['threshold_value']])
    
    print('Iteration: {}'.format(i), end='\r')


#append to dataframe
config_prod_df['speed']=speed
config_prod_df['speed_type']=speed_type
config_prod_df['brg_type']=brg_type
config_prod_df['speed_range']=speed_range
config_prod_df['threshold_X']=threshold_X
config_prod_df['threshold_Y']=threshold_Y
config_prod_df['threshold_Z']=threshold_Z
config_prod_df['temperature']=temperature_
#config_prod_df['audio']=audio_
config_prod_df['acceleration']=acceleration_
config_prod_df['lor']=lor
config_prod_df['fmax']=fmax
config_prod_df['fmin']=fmin


connection = psycopg2.connect(user="iuprod",
                                          password="pratiti",
                                          host="10.0.1.253",
                                          port="5432",
                                          database="iu_admin")

cursor = connection.cursor()
query = f"""
    select mo.label as monitor_label,mo.id as monitor_id, m.label as machine_name, m.id as machine_id, p.name as plant_name, o.name as org_name
from monitors mo
inner join machines m on mo.machine_id = m.id
inner join machinegroups mg on m.machine_group_id = mg.id
inner join plants p on p.id = mg.plant_id
inner join organizations o on o.id = p.organization_id
order by machine_id, monitor_id;
    """
cursor.execute(query)
records = cursor.fetchall()
result = []

for row in records:
    data = {}
    data['monitor_name'] = row[0]
    data['monitor_id'] = row[1]
    data['machine_name'] = row[2]
    data['machine_id'] = row[3]
    data['plant_name'] = row[4]
    data['org_name'] = row[5]
   
    result.append(data)

prod_db=pd.DataFrame(result)
master_prod_df=config_prod_df.merge(prod_db,how='inner',on='monitor_id')

master_prod_df.to_excel('configuration_data_.xlsx')
```
